LookupError Exception is the Base class for errors raised when something can't be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError. An IndexError is raised when a sequence reference is out of range.

```python
#Example of KeyError
try :
    d = {'key' :'aleem',1: [1,2,3,4,5]}
    print(d['key1'])
except KeyError as e :
    print(e)
```

Q.6 -List down some best practices for exception handling in python.

Here are some best practices for exception handling in Python:

Be specific with exception handling: Catch specific exceptions rather than using a broad except block. This helps in better understanding and handling of different types of exceptions.

Use multiple except blocks: If you need to handle different exceptions differently, use multiple except blocks to handle each exception separately. This allows you to provide specific error handling for different types of exceptions.

Use finally block for cleanup: If you have resources that need to be released or cleanup operations to be performed, use the finally block. This block is executed regardless of whether an exception occurred or not.

Avoid bare except: Avoid using a bare except block (except:) as it catches all exceptions, including system exit exceptions, which may hide important errors. Instead, catch specific exceptions or use Exception as a base class.
